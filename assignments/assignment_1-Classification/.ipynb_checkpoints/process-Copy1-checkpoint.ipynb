```python
import numpy as np
import pandas as pd
from dt_analysis import DecisionTreeC_Analysis
from xgb_analysis import XGBC_Analysis
from mlpc_analysis import MLPC_Analysis
from svm_analysis import SVMC_Analysis
from knn_analysis import KNNC_Analysis
```

```python
df_student = pd.read_csv("./../../data/cleaned_student_data.csv")
df_student.index.name = 'Student Dataset'
# df_housing = pd.read_csv("./../../data/cleaned_housing_data_preped_class_nozero.csv")
# df_housing.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'OBJECTID', '2018-07-22T18:01:43.000Z'], axis=1, inplace=True)
# df_housing.drop(columns=[i for i in df_housing.columns if i.find('Police') != -1], axis=1, inplace=True)
# df_housing.index.name = 'Housing Dataset'

# df_housing = df_housing.sample(30000, random_state=7308)

datasets = [df_student]
targets = ['gender']
```

```python
for dataset, target in zip(datasets, targets):
    analysis = DecisionTreeC_Analysis(dataset=dataset, target=target, save=False)
    analysis.general_analysis()
    r1 = analysis.depth_analysis(range_=np.linspace(2, 28, 18, dtype=int))
    r2 = analysis.max_node_analysis(range_=np.linspace(2, 500, 20, dtype=int))
    r3 = analysis.min_samples_analysis(range_=range(1,21))
    
    analysis = XGBC_Analysis(dataset=dataset, target=target, save=False)
    analysis.general_analysis()
    r3 = analysis.depth_analysis(range_=range(2,20))
    r4 = analysis.n_estimator_analysis()
    r5 = analysis.lr_analysis()
    
    analysis = SVMC_Analysis(dataset=dataset, target=target, save=False)
    analysis.general_analysis()
    r6 = analysis.kernel_analysis()
    r7 = analysis.penalty_analysis()
    
    analysis = MLPC_Analysis(dataset=dataset, target=target, save=False)
    analysis.general_analysis()
    r8 = analysis.max_iteration_analysis()
    r8 = analysis.hidden_layer_analysis()
    r9 = analysis.activation_analysis()
    
    analysis = KNNC_Analysis(dataset=dataset, target=target, save=True)
    analysis.general_analysis()
    r10 = analysis.n_neighbors_analysis()
```
