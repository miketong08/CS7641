```python
b_codes = [i.split('_')[0] for i in df_residential.columns if i.endswith('_D')]
```

```python
df_residential_drop_col = ['OBJECTID', 'QUALIFIED', 'USECODE', 'GIS_LAST_MOD_DTTM', 'SALEDATE']
df_residential_drop_col.extend(b_codes)
```

```python
df_residential_c = df_residential.drop(columns=df_residential_drop_col)
df_residential_c.shape
```

```python
df_residential_c[['YR_RMDL', 'PRICE']] = df_residential_c[['YR_RMDL', 'PRICE']].fillna(value=-1)
```

```python
df_main = df_residential_c.join(df_address, lsuffix='SSL', rsuffix='SSL')
df_main.shape
```

```python
df_main.sample(2)
```

```python
df_main.columns
```

```python
df_main.isna().sum().sort_values(ascending=False)
```

```python
cols = list(df_residential_c.columns)
cols.append('VOTE_PRCNCT')
cols.remove('SSL')
df_t = df_main[cols]
```

```python
df_t.columns
```

```python
encode = [i for i,v in df_t.dtypes.iteritems() if v == object]
encode.remove('VOTE_PRCNCT')
```

```python
encode
```

```python
dummies = pd.concat([pd.get_dummies(df_t[col]) for col in encode], axis=1)
df_t = pd.concat([df_t, dummies], axis=1)
df_t.drop(columns=encode, inplace=True)
df_t = df_t.dropna()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
x = df_t.drop(columns=['VOTE_PRCNCT'])
y = df_t['VOTE_PRCNCT']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

preds = clf.predict(x_test)
print(classification_report(y_test, preds))
```

```python
pd.DataFrame(list(zip(x.columns, clf.feature_importances_)), index=x.columns).drop(columns=0)[1].sort_values(ascending=False)
```
