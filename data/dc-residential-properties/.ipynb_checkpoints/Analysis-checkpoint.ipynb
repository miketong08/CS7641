```python
import pandas as pd
```

```python
df = pd.read_csv('./../cleaned_housing_data.csv')
```

```python
thirds = df.shape[0]/3 # locations to split data evenly into thirds
df_first_split = df['PRICE'].sort_values().iloc[int(thirds)]
df_second_split = df['PRICE'].sort_values().iloc[int(thirds * 2)]
```

```python
df['class'] = 'err'
```

```python
print("Group 1 is lte to {}".format(int(df_first_split)))
print("Group 2 is gt {} and lte to {}".format(int(df_first_split), int(df_second_split)))
print("Group 3 is gt {}".format(int(df_second_split)))

for i,v in df.iterrows():
    if v['PRICE'] <= df_first_split:
        df.loc[i, 'class'] = 'Group 1'
    if v['PRICE'] > df_first_split and v['PRICE'] <= df_second_split:
        df.loc[i, 'class'] = 'Group 2'
    if v['PRICE'] > df_second_split:
        df.loc[i,'class'] = 'Group 3'

df.drop(columns='PRICE', inplace=True)
```

```python
from collections import defaultdict

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, recall_score, f1_score

```

```python
X = df.drop(columns=('class'), axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = DecisionTreeClassifier(random_state=7641, criterion='entropy')
clf.fit(X_train, y_train)

preds = clf.predict(X_test)
print(classification_report(y_pred=preds, y_true=y_test))
```

```python
accuracy_score(y_pred=preds, y_true=y_test)
```

```python
def prep_data_for_DT(df, target, test_size=0.2, random_state=7308):
    X = df.drop(columns=(target), axis=1)
    y = df[target]

    return train_test_split(X, y, test_size=test_size, random_state=random_state)      
 
def prep_housing(df, verbose=True):
    df_c = df.copy()
    thirds = df.shape[0]/3 # locations to split data evenly into thirds
    df_first_split = df['PRICE'].sort_values().iloc[int(thirds)]
    df_second_split = df['PRICE'].sort_values().iloc[int(thirds * 2)]

    df_c['class'] = 'err'
    if verbose:
            print("Group 1 is lte to {}".format(int(df_first_split)))
            print("Group 2 is gt {} and lte to {}".format(int(df_first_split), int(df_second_split)))
            print("Group 3 is gt {}".format(int(df_second_split)))
            print("Adding `class` column via Group designations, this may take a while")

    for i,v in df_c.iterrows():
        if v['PRICE'] <= df_first_split:
            df_c.loc[i, 'class'] = 'Group 1'
        if v['PRICE'] > df_first_split and v['PRICE'] <= df_second_split:
            df_c.loc[i, 'class'] = 'Group 2'
        if v['PRICE'] > df_second_split:
            df_c.loc[i,'class'] = 'Group 3'

    return df_c.drop(columns='PRICE')
```

```python
def clean_student_data(DATA_PATH, target='gender', save_csv=False):
	"""
	"""

	df_m = pd.read_csv(DATA_PATH)
	df_categorical = [i for i,v in df_m.dtypes.iteritems() if v==object]
	df_categorical.remove(target)
	df_m = create_dummies(df_m, df_categorical)

	if type(save_csv) == str:
		df_m.to_csv(save_csv)

	return df_m

def create_dummies(df, cols):
	df_dummies = pd.concat([pd.get_dummies(df[col]) for col in cols], axis=1)
	df_c = pd.concat([df, df_dummies], axis=1)
	df_c.drop(columns=cols, axis=1, inplace=True)

	return df_c
```

```python
def plot_depth(cols, metric, prune, target, save=False):
    plt.gcf().set_size_inches(8, 5)

    for col in cols:
        plt.plot(range(2,20), prune[col])

    plt.legend(['_'.join(i.split('_')[1:]) for i in cols], ncol=1, loc=4)
    plt.xlabel('Max Depth')
    plt.ylabel('{} Score (Weighted)'.format(metric.capitalize()))
    if target=='gender':
        plt.title('{} of Train and Test for {}'.format(metric.capitalize(), target.capitalize()))
    if target=='class':
        plt.title('{} of Train and Test for {}'.format(metric.capitalize(), target.capitalize()))
        
    plt.show()
```

```python
df_student = pd.read_csv("./../cleaned_student_data.csv")
# df_housing = pd.read_csv("./../cleaned_housing_data.csv")
# df_housing = prep_housing(df_housing, verbose=False)

df_student_target = 'gender'
df_housing_target = 'class'

for dataset, target in zip([df_student], [df_student_target]):
    X_train, X_test, y_train, y_test = prep_data_for_DT(df=dataset, target=target,
                                                        random_state=7308)
    
    # depth analysis
    depth_d = defaultdict(list)

    for d in range(2, 20):
        clf = DecisionTreeClassifier(max_depth=d, random_state=7308)
        clf.fit(X_train, y_train)
        preds_train = clf.predict(X_train)
        preds_test = clf.predict(X_test)

        depth_d['acc_train_gini'].append(accuracy_score(y_true=y_train, y_pred=preds_train))
        depth_d['f1_train_gini'].append(f1_score(y_true=y_train, y_pred=preds_train, average='weighted'))

        depth_d['acc_test_gini'].append(accuracy_score(y_true=y_test, y_pred=preds_test))
        depth_d['f1_test_gini'].append(f1_score(y_true=y_test, y_pred=preds_test, average='weighted'))
        
        clf = DecisionTreeClassifier(criterion='entropy', max_depth=d, random_state=7308)
        clf.fit(X_train, y_train)
        preds_train = clf.predict(X_train)
        preds_test = clf.predict(X_test)

        depth_d['acc_train_entropy'].append(accuracy_score(y_true=y_train, y_pred=preds_train))
        depth_d['f1_train_entropy'].append(f1_score(y_true=y_train, y_pred=preds_train, average='weighted'))

        depth_d['acc_test_entropy'].append(accuracy_score(y_true=y_test, y_pred=preds_test))
        depth_d['f1_test_entropy'].append(f1_score(y_true=y_test, y_pred=preds_test, average='weighted'))

    min_samples = defaultdict(list)
    
    for m in range(2,100):
        clf = DecisionTreeClassifier(max_leaf_nodes=m, random_state=7308)
        clf.fit(X_train, y_train)
        preds_train = clf.predict(X_train)
        preds_test = clf.predict(X_test)

        min_samples['acc_train_gini'].append(accuracy_score(y_true=y_train, y_pred=preds_train))
        min_samples['f1_train_gini'].append(f1_score(y_true=y_train, y_pred=preds_train, average='weighted'))

        min_samples['acc_test_gini'].append(accuracy_score(y_true=y_test, y_pred=preds_test))
        min_samples['f1_test_gini'].append(f1_score(y_true=y_test, y_pred=preds_test, average='weighted'))
        
        clf = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=m, random_state=7308)
        clf.fit(X_train, y_train)
        preds_train = clf.predict(X_train)
        preds_test = clf.predict(X_test)

        min_samples['acc_train_entropy'].append(accuracy_score(y_true=y_train, y_pred=preds_train))
        min_samples['f1_train_entropy'].append(f1_score(y_true=y_train, y_pred=preds_train, average='weighted'))

        min_samples['acc_test_entropy'].append(accuracy_score(y_true=y_test, y_pred=preds_test))
        min_samples['f1_test_entropy'].append(f1_score(y_true=y_test, y_pred=preds_test, average='weighted'))
    
    # plotting 
    
    plt.gcf().set_size_inches(8, 5)
    accs = ['acc_train_gini', 'acc_train_entropy', 'acc_test_gini', 'acc_test_entropy']
    f1s = ['f1_train_gini', 'f1_train_entropy', 'f1_test_gini', 'f1_test_entropy']

    plot_depth(metric='accuracy', target=target, cols=accs)
    plot_depth(metric='f1', target=target, cols=f1s)
```
